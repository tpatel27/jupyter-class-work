### Defining axes, ticks, and grids

```python
# https://www.youtube.com/watch?v=D8rwqYcVvmM
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
# https://stackoverflow.com/questions/43326680/what-are-the-differences-between-add-axes-and-add-subplot
ax = fig.add_axes([.1,.1,1,1])

ax.plot(x,y)
```

```python
# https://www.youtube.com/watch?v=xcFrvlKbWVU
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

### Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()
# fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
#     ncols=2,
#     nrows=2,
#     figsize=(2, 2))

fig, axes = plt.subplots(
    ncols=2,
    nrows=2,
    figsize=(2, 2))

ax1, ax2, ax3, ax4 = axes.flatten()

ax1.plot(x)
ax2.plot(x,y)
ax3.plot(y)
ax4.plot(y,x)
```

```python
#alternate way for above plot
fig1 = plt.figure()
fig1, ax = plt.subplots(2,2)

ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,3)
ax4 = plt.subplot(2,2,4)

ax1.plot(x)
ax2.plot(x,y)
ax3.plot(y)
ax4.plot(y,x)
```
