## Segment 4 - Concatenating and transforming data

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
DF_obj1=pd.DataFrame(np.arange(36).reshape(6,6))
DF_obj1
```

```python
DF_obj2=pd.DataFrame(np.arange(15).reshape(5,3))
DF_obj2
```

### Concatenating data

```python
pd.concat([DF_obj1, DF_obj2], axis=1)
```

```python
out=pd.concat([DF_obj1, DF_obj2])
out
```

#### Dropping data
