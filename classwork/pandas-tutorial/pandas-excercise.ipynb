# Making Changes to Dataset

```python
#To get the total in a new column
df_csv['Total'] = df_csv['HP'] + df_csv['Attack'] + df_csv['Defense'] + df_csv['Sp. Atk'] + df_csv['Sp. Def'] + df_csv['Speed']

df_csv.head(5)
```

```python
#To get the total using different way
df_csv['Total 2'] = df_csv.iloc[:, 2:8].sum(axis=1)
#in the above code ':' is used to get all rows and the next '4:9' gives rows from pos 4 to 9 '.sum' gives the total and axis=1 is used to give the sum horizontally, 0 to get vertically.
df_csv.head(5)
```
